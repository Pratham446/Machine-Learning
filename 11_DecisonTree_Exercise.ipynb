```python
import pandas as pd
df=pd.read_csv('titanic (1).csv')
df.head()
```

```python
input=df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked','Survived'],axis='columns')
input
```

```python
target=df.Survived
target.head(10)
```

```python
input.Sex = input.Sex.map({'male': 1, 'female': 2})
input
```

```python
input.Age[:10]
```

```python
# Age contains NA column so fill it with mean value 
```

```python
input.Age=input.Age.fillna(input.Age.mean())
input.head()
```

```python
# time to split the data for training and testing 
```

```python
target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(input,target,test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
# now time for implementing decsion tree algorithm 
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
